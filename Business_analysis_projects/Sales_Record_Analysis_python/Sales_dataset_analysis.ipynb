# Pivot Table Analysis.

In this section, we will create pivot tables using the available columns in the dataset to facilitate a more in-depth analysis. These pivot tables will allow us to summarize, aggregate, and explore the data from various perspectives, enabling a deeper understanding of the sales patterns and trends within the dataset.

```python
pivot_table_1 = pd.pivot_table(data=dataset, values='Units Sold', index='Region', columns=['Order Priority'])
pivot_table_1
```

```python
pivot_table_1.plot(kind='line',figsize=(14,5))
plt.suptitle('Comparison between order priority to Region.',fontsize=20)
plt.grid()
```

As per the Pivot_table 1 analysis and visualization reveal distinct patterns in order priorities across different regions. The Asia region stands out with the highest number of Critical orders, while Australia and Oceania lead in High priority orders. North America, on the other hand, dominates in both Medium and Low priority orders. These regional variations in order priorities suggest potential differences in customer needs, market dynamics, or operational practices across these geographic areas.

```python
pivot_table_2=pd.pivot_table(data=dataset,values='Units Sold',index='Region',columns=['Sales Channel'])
pivot_table_2
```

```python
pivot_table_2.plot(kind='bar',figsize=(15,5))
plt.suptitle('Sales channel the order made as per the Region',fontsize=20)
```

Analysis of pivot table 2, focusing on sales channels across regions, reveals distinct preferences in ordering methods. Australia and Oceania demonstrate the highest proportion of online orders, while North America leads in offline orders. These findings highlight regional differences in consumer behavior and sales channel effectiveness, which could inform targeted marketing and distribution strategies.

```python
pivot_table_3 = pd.pivot_table(data=dataset, values='Units Sold', index='Item Type', columns=['Region'])
pivot_table_3
```

```python
pivot_table_3.plot(kind='line',figsize=(15,5))
plt.suptitle('Item Type sold accross Region',fontsize=20)
plt.grid()
```

Analysis of Pivot Table 3 reveals distinct regional preferences for various product categories. Key findings include: Baby Foods are most popular in the Middle East and North Africa; Beverages and Clothes lead in North America; Asia shows high demand for Cereal, Snacks, and Vegetables; Central America and the Caribbean favor Fruits and Meat; Office Supplies are top-selling in the Middle East and North Africa; and Personal Care products are most sought after in Australia and Oceania. These insights highlight regional variations in consumer preferences and purchasing patterns across different product categories.

# Sales performance.

```python
# Sales Trends Over Time
dataset['Order Date'] = pd.to_datetime(dataset['Order Date'])
sales_trends = dataset.groupby(dataset['Order Date'].dt.to_period('M'))['Total Revenue'].sum().reset_index()
sales_trends['Order Date'] = sales_trends['Order Date'].dt.to_timestamp()
```

```python
plt.figure(figsize=(14, 7))
dataset.groupby('Order Date')['Units Sold'].sum().plot()
plt.suptitle('Units Sold Over Time',fontsize=20)
plt.grid()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=sales_trends, x='Order Date', y='Total Revenue', marker='o')
plt.title('Sales Trends Over Time')
plt.ylabel('Total Revenue')
plt.xlabel('Order Date')
plt.xticks(rotation=45)
plt.grid()
```

Based on the chart above, we can clearly observe that sales performance significantly improved during the 2013-2014 times and goes down at the time period of 2011 & 2017.

# Feature Engineer

```python
priority_mapping = {'Low': 1, 'Medium': 2, 'High': 3, 'Critical': 4}
dataset['Order_Priority_Encoded'] = dataset['Order Priority'].map(priority_mapping)
```

```python
dataset['Delivery_Days'] = (pd.to_datetime(dataset['Ship Date']) - pd.to_datetime(dataset['Order Date'])).dt.days
```

```python
dataset['Profit_Margin'] = dataset['Total Profit'] / dataset['Total Revenue']
```

```python
dataset['Units_Price_Interaction'] = dataset['Units Sold'] * dataset['Unit Price']
```
